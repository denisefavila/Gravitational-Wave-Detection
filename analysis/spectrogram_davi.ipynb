```python
# arquivos disponiveis:
# 00000e74ad.npy (target=1) 00007a006a.npy (target=0) 0000c3b9c9.npy (target=0)
# 00010beb4a.npy (target=1) 00014b7a9d.npy (target=1) 0001808ecc.npy (target=1)
# 00001f4945.npy (target=1) 0000a38978.npy (target=1) 0000d61b7b.npy (target=1)
# 000118b40d.npy (target=0) 0001616241.npy (target=0) 0000661522.npy (target=0)
# 0000bb9f3e.npy (target=1) 0001016d12.npy (target=1) 0001388506.npy (target=1)
# 00017d3cf3.npy (target=1)

import numpy as np
import matplotlib.pyplot as plt
from gwpy.timeseries import TimeSeries
from gwpy.time import to_gps

file = '00017d3cf3.npy'
path = f'../data/{file}'

signals = np.load(path)

size = 2**12
gps = to_gps('Apr 21 1991 02:30:45.391')
sample_rate = 2**11

qrange = (16, 32) # padrao = (4, 64)
frange = (16, 256) # padrao automatico

plt.figure(figsize=(20,5))

plt.plot(signals[0,:], c="firebrick", label="detector 1")
plt.plot(signals[1,:], c="mediumseagreen", label="detector 2")
plt.plot(signals[2,:], c="slateblue", label="detector 3")
plt.title(f'Example_{file}');
plt.legend();
```

```python
time_series = TimeSeries(signals[0, :], t0=gps, sample_rate=sample_rate, name= 'strain')

plot = time_series.q_transform(qrange, frange).plot()
ax = plot.gca()
ax.set_epoch(gps)
ax.set_yscale('log')
ax.colorbar(label="Normalised energy")
#plot.colorbars[0].mappable.set_clim(0,20)
```
