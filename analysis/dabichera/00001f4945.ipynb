```python
import numpy as np

file = '00001f4945.npy'

data = np.load(file)
```

```python
from gwpy.timeseries import TimeSeries
from gwpy.time import to_gps

gps = to_gps('Apr 21 1991 02:30:45.391')
```

```python
dect1 = TimeSeries(data[0], t0=gps, sample_rate=2**11, name='Strain')

spec1 = dect1.q_transform(frange=(30, 645), qrange=(30, 50))

plot = spec1.plot()
ax = plot.gca()
ax.set_epoch(gps)
ax.set_yscale('log')
ax.colorbar(label="Normalised energy")
plot.colorbars[0].mappable.set_clim(0,20)
```
