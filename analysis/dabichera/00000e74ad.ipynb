```python
import numpy as np

file = '00000e74ad.npy'

data = np.load(file)
```

```python
from gwpy.timeseries import TimeSeries
from gwpy.plot import Plot
from gwpy.time import to_gps

sign = [TimeSeries(data[i], unit='strain', t0=to_gps('Apr 21 1991 02:30:45.391'), sample_rate=2**11)
        for i in range(3)]

plot = Plot(*sign, separate=True, sharex=True)
ax = plot.gca()
```

```python
hq = sign[0].q_transform()

plot = hq.plot()
```
